```sql
CREATE SCHEMA IF NOT EXISTS project;
```

```sql
CREATE OR REPLACE TABLE project.customer_behaviour AS
SELECT
  `Customer ID`           AS customer_id,
  Age                     AS age,
  Gender                  AS gender,
  `Item Purchased`        AS item_purchased,
  Category                AS category,
  `Purchase Amount (USD)` AS purchase_amount_usd,
  Location                AS location,
  Season                  AS season,
  `Review Rating`         AS review_rating,
  `Shipping Type`         AS shipping_type,
  `Previous Purchases`    AS previous_purchases,
  `Payment Method`        AS payment_method
FROM main.default.customer_behaviour;
```

```sql
SELECT *
FROM project.customer_behaviour
LIMIT 10;
```

```sql
SELECT season,
       ROUND(AVG(purchase_amount_usd), 2) AS avg_spend
FROM project.customer_behaviour_final
GROUP BY season
ORDER BY avg_spend DESC;
```

```sql
SELECT age_group,
       ROUND(AVG(purchase_amount_usd), 2) AS avg_spend
FROM project.customer_behaviour_final
GROUP BY age_group
ORDER BY avg_spend DESC;
```
